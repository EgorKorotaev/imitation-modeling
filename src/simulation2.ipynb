```python
from simulation import Simulation

def experiment(EXPERIMENTS=100, TOTAL_PACKETS=500, FAILOVER_THRESHOLD=20):
    all_results = []

    for experiment_id in range(EXPERIMENTS):
        simulation = Simulation(total_packets=TOTAL_PACKETS, reserved_system_threshold=FAILOVER_THRESHOLD)
        simulation.run_simulation()
        all_results.append(simulation.experiment_results)



    failover_system_turned_on = len(list(filter(lambda x: x is True, map(lambda x: x[0], all_results))))

    print(f'Резервная система включилась в {100 * failover_system_turned_on / EXPERIMENTS}% случаев')


    import matplotlib.pyplot as plt
    %matplotlib inline
    plt.rcParams.update({'figure.figsize':(7,5), 'figure.dpi':100})

    # Plot Histogram on x
    x = list(map(lambda a: a[1], all_results))
    plt.hist(x, bins=50)
    plt.gca().set(title='Потерянные пакеты', ylabel='Частота');
```

```python
experiment(100, 500, 20)
```

```python
experiment(100, 500, -1)
```

```python
experiment(100, 500, 23)
```
